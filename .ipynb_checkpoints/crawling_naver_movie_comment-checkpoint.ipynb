```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd

def get_review(url) :
    resp = requests.get(url)
    html = BeautifulSoup(resp.content, 'html.parser')
        
    score_result = html.find('div', {'class': 'score_result'})
    reviews = score_result.findAll('li')
    
    for review in reviews :
        review_score = review.find('em').getText()
        if(review_score == input_score) :
            try : 
                review_nickname = review.findAll('a')[0].find('span').getText()
            except :
                review_nickname = review.findAll('a')[1].find('span').getText()
            review_text = review.find('p').getText().strip()
            review_date = datetime.strptime(review.find('dt').findAll('em')[-1].getText(), "%Y.%m.%d %H:%M")
            
            global reviewList
            reviewList.append([review_score, review_nickname, review_text, review_date])
            
            if len(reviewList) == 10 :
                df = pd.DataFrame(reviewList, columns=['평점', '닉네임', '영화평', '작성일'])
                df = df.style.set_properties(**{'text-align': 'left'})
                display(df)
                input_continue = input("계속 진행하시겠습니까? : ")
                if input_continue == 'n' or input_continue == 'N' : return False
                reviewList = []
                
    return True

reviewList = []
input_score = input("당신이 보고 싶은 평점을 입력하시오 : ")
page_num = 1
while True : 
    print(str(page_num) + "번째 페이지 크롤링 중...")
    input_continue = get_review("https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=37886&page=" + str(page_num))
    if input_continue == False : break
    page_num += 1
```
